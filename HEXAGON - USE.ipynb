```python
# ALRIGHT THIS IS THE ACTUAL CALCULATION
# DIMENSIONS
# DISTANCE BETWEEN TUBES 
# RADIUS

# CHECK TO MAKE SURE INPUT HAS NUMBERS 

import numpy as np
import matplotlib.pyplot as plt

def value_check(x):
    values = []
    for item in x:
        if ord(item) > 57:
            values.append(False)
        elif ord(item) < 45:
            values.append(False)
        elif ord(item) == 47:
            values.append(False)
        else:
            values.append(True)
    if False in values:
        print("No alpha characters!")
        return(False)
    else: 
        return(True)
    
# X DIMENSION
        
x_dimension = input("What is the length of the system?: ")
val = value_check(x_dimension)

while val == False:
    x_dimension = input("What is the length of the system?: ")
    val = value_check(x_dimension)
    if val == True:
        x_dim = float(x_dimension)
if val == True:
    x_dim = float(x_dimension)

# Y DIMENSION

y_dimension = input("What is the height of the system?: ")
val1 = value_check(y_dimension)

while val1 == False:
    y_dimension = input("What is the height of the system?: ")
    val = value_check(y_dimension)
if val == True:
    y_dim = float(y_dimension)
    
print("", x_dim, "cm long","\n", y_dim, "cm tall")

# RADIUS

radius = input("What is the radius of the tubes?: ")
val2 = value_check(radius)

while val2 == False:
    radius = input("What is the radius of the system?: ")
    val = value_check(radius)
if val == True:
    radius_cm = float(radius)
    
# THICKNESS OF THE TUBE

tube_thick = input("What is the thickness of the tubes?: ")
val3 = value_check(tube_thick)

while val3 == False:
    tube_thick = input("What is the thickness of the tubes?: ")
    val = value_check(tube_thick)
if val == True:
    tube_thick_cm = float(tube_thick)
    
# DISTANCE APART

distance = input("What distance do you want the tubes apart? : ")
val4 = value_check(distance )

while val4 == False:
    distance= input("What is the thickness of the tubes?: ")
    val = value_check(distance)
if val == True:
    distance_cm = float(distance)
    
print("", radius_cm, "cm radius\n", tube_thick_cm, "cm thickness\n", distance_cm, "cm apart")
```

```python


# center1 =x: leftover from dimensions/2 + tube_thick_radius_cm
# center2 =x: leftover from deimensions/2 +tube_thick_radiu_cm
#centroid = x: sqrt(3)*distance apart from each other y: equals .5 distance apart from each other
#center 6 = x = sqrt(3)* distance apart from each other (same as center) y: centroid+distance apart from each other
#center 3 = same x and center 6, y = centroid - total distance apart
#center 4 = 2* sqrt(distance apart from each other), y= same as center 2
#center 5 = same as center 4, y = same as center 1

#distance from top to bottom is: TOTAL D + distance apart + TOTAL D (2*r + 2* thickness)+distance apart + TOTAL D
#distance from side to side is: 2*sqrt(3) distance apart (with radial component included) + D (D/2 + D/2)

# from here calculate the dimensions of hexagon
radial = radius_cm + tube_thick_cm

center_distance = (2*tube_thick_cm) + (2*radius_cm) + distance_cm
print(center_distance)

total_x_distance = (2*tube_thick_cm) + (2*radius_cm) + (np.sqrt(3)*(center_distance))
x_multiplying_distance = (tube_thick_cm) + (radius_cm) + (np.sqrt(3)*(center_distance))
print(total_x_distance)
print(x_multiplying_distance)

total_y_distance = (2*tube_thick_cm) + (2* radius_cm) + (2*center_distance)
y_multiplying_distance = (tube_thick_cm) + (radius_cm) + (2*center_distance)
print(total_y_distance, "y")
print(y_multiplying_distance, "y")

x_hexes = x_dim/x_multiplying_distance
x_int = int(x_hexes)
print(x_int, "the number of times the x-axis can fit")
#need even and odd calculations for this and y distance!

x_edge_val = ((x_dim) - ((x_int*(np.sqrt(3)*center_distance)+(2*radial))))/2
print(x_edge_val, "cm of poly on the x sides")

y_hexes = y_dim/y_multiplying_distance
y_int = int(y_hexes)
y_edge_val = ((y_dim) - ((y_int*2*center_distance)+(2*radial)))/2
print(y_edge_val, "cm of poly on the y sides")
print(y_int, "the number of times the y-axis can fit")


# center 1
c1x = round((x_edge_val + radial), 4)
c1y = round((y_edge_val + radial + (center_distance/2)), 4)

# center 2
c2x = round((c1x + (np.sqrt(3)*(center_distance/2))), 4)
c2y = round((radial + y_edge_val), 4)

# center 3
c3x = round((c1x + (2*np.sqrt(3)*(center_distance/2))), 4)
c3y = c1y

#center 4
c4x = c3x
c4y = round((c3y + center_distance ), 4)

#center 5
c5x = c2x
c5y = round((c4y + (center_distance/2)), 4)

#center 6
c6x = c1x
c6y = round((c1y + center_distance), 4)

#centroid
center_x = c2x
center_y = round((c1y + (center_distance/2)), 4)

#additional center (left corner)

acx = c1x
acy = c1y - center_distance

x = [c1x, c2x, c3x, c4x, c5x, c6x, center_x]
y = [c1y, c2y, c3y, c4y, c5y, c6y, center_y]

print(x, y)
```

```python
# Additional centers IN X DIMENSION
ac_oldx = [acx]
ac_oldy = [acy]

ac_newx = []
ac_newy = []

bint = x_int

def ac_circle(lista, listb):
    for item in lista:
        ac_int = bint 
        while ac_int > 0:
            item1 = round((item+(2*ac_int*np.sqrt(3)*(center_distance/2))), 4)
            ac_newx.append(item1)
            ac_int -= 1
    for item in listb:
        ac1_int = bint 
        while ac1_int > 0:
            item1 = round(item, 4)
            ac_newy.append(item1)
            ac1_int -= 1

ac_circle(ac_oldx, ac_oldy)

print(ac_newx, ac_newy)
print("Added additional centers, x coords", ac_newx, "y coords", ac_newy)

ac_oldx.extend(ac_newx)
ac_oldy.extend(ac_newy)

# print(ac_oldx, ac_oldy)
```

```python
# Additional centers IN Y DIMENSION
ac1_oldx = [acx]
ac1_oldy = [acy]

print(ac1_oldx, ac1_oldy)

ac1_newx = []
ac1_newy = []

bint = x_int + 1
print(bint)

def ac_circle(lista, listb):
    for item in lista:
        ac_int = bint - 1  
        while ac_int > 0:
            item1 = round((item+(2*ac_int*np.sqrt(3)*(center_distance/2))), 4)
            ac1_newx.append(item1)
            ac_int -= 1
    for item in listb:
        item1 = round((item+(1*(bint + 1)*center_distance)), 4)
        ac1_newy.append(item1)

ac_circle(ac1_oldx, ac_newy)


print(ac1_oldx, ac1_oldy)

ac1_oldy[0] = ac1_oldy[0] + ((bint+1)*center_distance)

print("Added additional centers, x coords", ac1_newx, "y coords", ac1_newy)

ac1_oldx.extend(ac1_newx)
ac1_oldy.extend(ac1_newy)

print(ac1_oldx, ac1_oldy)
```

```python
new_x = []
new_y = []
#set up xy_gon like ac_circle-----CJT 2/13/2019

cint = x_int - 1

def xy_agon(lista, listb):
    for item in lista:
        xy_int = cint
        while xy_int > 0:
            item1 = round(item+(2*xy_int*np.sqrt(3)*(center_distance/2)), 4)
            new_x.append(item1)
            xy_int -=1
    for item in listb:
        xy1_int = cint
        while xy1_int > 0:
            item1 = item
            new_y.append(item1)
            xy1_int -=1

if x_int > 1:
    xy_agon(x,y)

print(" Added x-coordinates", new_x, "\n Added y-coordinates", new_y)
print()

x.extend(new_x)
y.extend(new_y)

# print(x, y)

print(x,y)

print(len(x), len(y))
```

```python
new1_x = []
new1_y = []
#set up yx_gon like ac_circle-----CJT 2/13/2019

gint = y_int

def yx_agon(lista, listb):
    for item in listb:
        yx_int = gint 
        while yx_int > 1:
            item1 = round((item+(1*yx_int*center_distance)), 4)
            new1_y.append(item1)
            yx_int -=1
    for item in lista:
        yx1_int = gint
        while yx1_int > 1:
            item1 = item
            new1_x.append(item1)
            yx1_int -=1

if y_int > 1:
    yx_agon(x,y)

print(" Added x-coordinates", new1_x, "\n Added y-coordinates", new1_y)
print()

x.extend(new1_x)
y.extend(new1_y)

print(len(x), len(y))

x.extend(ac_oldx)
y.extend(ac_oldy)

print(len(x), len(y))

x.extend(ac1_oldx)
y.extend(ac1_oldy)

print(x,y)

print(len(x), len(y))
```
