```python
# ALRIGHT THIS IS THE ACTUAL CALCULATION
# DIMENSIONS
# DISTANCE BETWEEN TUBES 
# RADIUS

# CHECK TO MAKE SURE INPUT HAS NUMBERS 

import numpy as np
import matplotlib.pyplot as plt

def value_check(x):
    values = []
    for item in x:
        if ord(item) > 57:
            values.append(False)
        elif ord(item) < 45:
            values.append(False)
        elif ord(item) == 47:
            values.append(False)
        else:
            values.append(True)
    if False in values:
        print("No alpha characters!")
        return(False)
    else: 
        return(True)
    
# X DIMENSION
        
x_dimension = input("What is the length of the system?: ")
val = value_check(x_dimension)

while val == False:
    x_dimension = input("What is the length of the system?: ")
    val = value_check(x_dimension)
    if val == True:
        x_dim = float(x_dimension)
if val == True:
    x_dim = float(x_dimension)

# Y DIMENSION

y_dimension = input("What is the height of the system?: ")
val1 = value_check(y_dimension)

while val1 == False:
    y_dimension = input("What is the height of the system?: ")
    val1 = value_check(y_dimension)
if val1 == True:
    y_dim = float(y_dimension)
    
print("", x_dim, "cm long","\n", y_dim, "cm tall")

# RADIUS

radius = input("What is the radius of the tubes?: ")
val2 = value_check(radius)

while val2 == False:
    radius = input("What is the radius of the system?: ")
    val2 = value_check(radius)
if val2 == True:
    radius_cm = float(radius)
    
# THICKNESS OF THE TUBE

tube_thick = input("What is the thickness of the tubes?: ")
val3 = value_check(tube_thick)

while val3 == False:
    tube_thick = input("What is the thickness of the tubes?: ")
    val3 = value_check(tube_thick)
if val3 == True:
    tube_thick_cm = float(tube_thick)
    
# DISTANCE APART

distance = input("What distance do you want the tubes apart? : ")
val4 = value_check(distance )

while val4 == False:
    distance= input("What is the thickness of the tubes?: ")
    val4 = value_check(distance)
if val4 == True:
    distance_cm = float(distance)
```

```python


# from here calculate the dimensions of square
radial = radius_cm + tube_thick_cm

center_distance = (2*tube_thick_cm) + (2*radius_cm) + distance_cm
print(center_distance)


x_multiplying_distance = (radial)+ (center_distance)
print(x_multiplying_distance)


y_multiplying_distance = (radial) + (center_distance)
print(y_multiplying_distance, "y")

x_hexes = x_dim/x_multiplying_distance
x_int = int(x_hexes)
print(x_int, "the number of times the x-axis can fit")


x_edge_val = ((x_dim) - ((x_int*center_distance)+(2*radial)))/2
print(x_edge_val, "cm of poly on the x sides")

y_hexes = y_dim/y_multiplying_distance
y_int = int(y_hexes)
print(y_int, "the number of times the y-axis can fit")

y_edge_val = ((y_dim) - ((y_int*center_distance)+(2*radial)))/2

if y_edge_val < 0 :
    while y_edge_val < 0 :
        y_edge_val = ((y_dim) - (((y_int-1)*np.sqrt(3)*center_distance)+(2*radial)))/2
        y_int -= 1
else:
    pass

           
print(y_edge_val, "cm of poly on the y sides")


# center 1
c1x = round((x_edge_val + radial), 4)
c1y = round((y_edge_val + radial), 4)

# center 2
c2x = round((c1x), 4)
c2y = round((c1y + (center_distance)), 4)


x = [c1x, c2x]
y = [c1y, c2y]

print(x, y)
```

```python
# Additional centers IN X DIMENSION for C1
c1_x = [c1x]
c1_y = [c1y]

c1_newx = []
c1_newy = []

bint = x_int

def ac_circle(lista, listb):
    for item in lista:
        ac_int = bint 
        while ac_int > 0:
            item1 = round((item+(ac_int*center_distance)), 4)
            c1_newx.append(item1)
            ac_int -= 1
    for item in listb:
        ac1_int = bint 
        while ac1_int > 0:
            item1 = round(item, 4)
            c1_newy.append(item1)
            ac1_int -= 1

ac_circle(c1_x, c1_y)

print(c1_newx, c1_newy)
print("Added additional centers, x coords", c1_newx, "y coords", c1_newy)

c1_x.extend(c1_newx)
c1_y.extend(c1_newy)

print(c1_y)
```

```python
# Additional centers IN X DIMENSION for C2
c2_x = [c2x]
c2_y = [c2y]

c2_newx = []
c2_newy = []

bint = x_int
print(bint)

def ac_circle(lista, listb):
    for item in lista:
        ac_int = bint
        while ac_int > 0:
            item1 = round((item+(ac_int*center_distance)), 4)
            c2_newx.append(item1)
            ac_int -= 1
    for item in listb:
        ac1_int = bint 
        while ac1_int > 0:
            item1 = round(item, 4)
            c2_newy.append(item1)
            ac1_int -= 1

ac_circle(c2_x, c2_y)

print(c2_newx, c2_newy)
print("Added additional centers, x coords", c2_newx, "y coords", c2_newy)

c2_x.extend(c2_newx)
c2_y.extend(c2_newy)
```

```python
# Additional centers IN Y DIMENSION FOR C2

ac2_newx = []
ac2_newy = []

print(c2_y)

if y_int % 2 == 0:
    bint = y_int - 2 
else: 
    bint = y_int - 1
    
print(bint)


def ac_circle(lista, listb):
    for item in lista:
        ac_int = bint 
        while ac_int > 1:
            item1 = round(item, 4)
            ac2_newx.append(item1)
            ac_int -= 1
    for item in listb:
        ac_int = bint 
        while ac_int > 1:
            item1 = round((item+(1*(ac_int)*center_distance)), 4)
            ac2_newy.append(item1)
            ac_int -= 1
# can use c1_x, because the xs are the same!
ac_circle(c2_x, c2_y)


# print("Added additional centers, x coords", ac1_newx, "y coords", ac1_newy)

print(ac2_newx)
print(ac2_newy)
print(len(ac2_newx))
print(len(ac2_newy))

# c1_oldx.extend(ac1_newx)
# c1_oldy.extend(ac1_newy)
```
